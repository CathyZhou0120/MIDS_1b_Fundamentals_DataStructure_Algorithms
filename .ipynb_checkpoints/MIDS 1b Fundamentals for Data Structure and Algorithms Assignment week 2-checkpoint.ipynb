#### Assignment: write a program that will help to identify the "winning strategy"

```python
def thewinningStratgy (n):
    if n-1 < 1 and n/2 < 1:
        return "cold"
    elif 1<= (n-1)<2 and 1<=(n/2)<2:
        return "hot"
    else:
        subtract = thewinningStratgy(n-1)
        divide = thewinningStratgy(n/2)
        if subtract == "cold" or divide == "cold":
            return "hot"
        else:
            return "cold"
```

```python
thewinningStratgy(3)
```
